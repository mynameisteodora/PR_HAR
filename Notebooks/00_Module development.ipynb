```python
fig = plt.figure(figsize=(15, 30))
fig.subplots_adjust(hspace=1, wspace=0.5)
fig.suptitle('Histograms for activity 0', fontsize=18, y=0.90)
fig.tight_layout()
counter = 1
subject_num = 0
subject_list = sorted(get_subject_names())
accel_axes = ['accel_x', 'accel_y', 'accel_z']
for i in range(1, 46):
    ax = fig.add_subplot(15, 3, i)
    mask = (df_all_0_correct['subject'] == subject_list[subject_num])
    plt.hist(df_all_0_correct[mask][accel_axes[counter-1]])
    
    ax.set_xlim((-1.5, 1.5))
    
    if i == 1:
        plt.title('accel_x')
    elif i == 2:
        plt.title('accel_y')
    elif i == 3:
        plt.title('accel_z')
    
    if counter == 1:
        ax.set_ylabel('Subject {}'.format(subject_num))
    
    counter += 1
    
    if counter == 4:
        counter = 1
        subject_num += 1
```
